```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
df= pd.read_csv("/content/our_dataset.csv")
df.head()
```

```python
X = df['col1']
Y  =df['col2']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train,Y_test = train_test_split(X, Y, test_size=0.2)

```

```python
Y_train
```

```python
train_df = pd.DataFrame(X_train)
train_df['target'] = Y_train

eval_df = pd.DataFrame(X_test)
eval_df['target'] = Y_test

train_df.shape, eval_df.shape
```

```python
train_df
```

```python
# We will import ClassificationModel - as we need to solve binary text classification
from simpletransformers.classification import ClassificationModel
import pandas as pd
import logging
import sklearn


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.INFO)


model = ClassificationModel('bert', 'Ammar-alhaj-ali/arabic-MARBERT-dialect-identification-city', use_cuda=False, args={'reprocess_input_data': True, 'overwrite_output_dir': True})
```

```python
model.train_model(train_df)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_df, acc=sklearn.metrics.accuracy_score)
```
